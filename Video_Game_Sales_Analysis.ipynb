```python
avg_sales_per_genre_in_na = video_game_sale.groupby('Genre')['North_America_Sales'].mean()
```

```python
avg_sales_per_genre_in_na
```

```python
avg_sales_per_genre_in_europe = video_game_sale.groupby('Genre')['Europe_Sales'].mean()
```

```python
avg_sales_per_genre_in_europe
```

```python
avg_sales_per_genre_in_japan = video_game_sale.groupby('Genre')['Japan_Sales'].mean()
```

```python
avg_sales_per_genre_in_japan
```

```python
avg_sales_per_genre_in_other = video_game_sale.groupby('Genre')['Other_Sales'].mean()
```

```python
avg_sales_per_genre_in_other
```

```python
avg_sales_df = pd.DataFrame({
    'North America': avg_sales_per_genre_in_na,
    'Europe': avg_sales_per_genre_in_europe,
    'Japan': avg_sales_per_genre_in_japan,
    'Other': avg_sales_per_genre_in_other,
})
```

```python
regions = ['North America', 'Europe', 'Japan', 'Other']

plt.figure(figsize=(15, 12))  

for i, region in enumerate(regions, start=1):
    plt.subplot(2, 2, i)  
    avg_sales_df[region].sort_values(ascending=False).plot(kind='bar', color='#6B3333')
    plt.title(f'Average Sales per Genre in {region}')
    plt.xlabel('Genre')
    plt.ylabel('Average Sales (millions)')
    plt.xticks(rotation=90)

plt.tight_layout()
plt.show()
```

1- North America (NA)
Top Genres:Platform (highest average sales) Shooter, Racing, and Sports
Lowest Performing Genres: Strategy and Adventure
Insight: NA shows a strong preference for Platform and Shooter games.
Action and Sports also hold significant market interest.
Strategic and Adventure genres appear to underperform on average.

2- Europe (EU)
Top Genres: Shooter (highest), followed closely by Platform Racing and Sports
Lowest Performing Genres: Strategy and Adventure
Insight: Shooters are the leading genre in Europe.
Preferences closely mirror North America, with a slight edge toward Shooter over Platform.
Again, Strategy and Adventure perform poorly.

3-Japan (JP)
Top Genres: Role-Playing (by far the most popular) Platform and Fighting
Lowest Performing Genres: Shooter and Adventure
Insight: Japan shows a unique trend, heavily favoring Role-Playing games.
Platform games are still strong, but shooters are least popular, showing a contrast with Western regions.
Cultural preferences in game types are clearly different.

4- Other Regions
Top Genres: Shooter and Racing Platform and Sports
Lowest Performing Genres: Strategy, Adventure, and Puzzle
Insight: Shooter games lead again, suggesting a global trend in their popularity.
Like NA and EU, Adventure and Strategy genres remain less attractive on average.

Overall Insights:Shooters and Platform games dominate in most regions except Japan.
Role-Playing games are extremely strong in Japan, showing a distinct regional preference.
Adventure and Strategy genres underperform globally, suggesting niche interest or lower market penetration.
Marketing and development should align genre focus with regional preferences to maximize sales impact.

```python
avg_sales_per_year_in_other = video_game_sale.groupby('Year')['Other_Sales'].mean()
```

```python
avg_sales_year_df = pd.DataFrame({
    'North America': avg_sales_per_year_in_na,
    'Europe': avg_sales_per_year_in_europe,
    'Japan': avg_sales_per_year_in_japan,
    'Other': avg_sales_per_year_in_other
})
```

```python
regions = ['North America', 'Europe', 'Japan', 'Other']

plt.figure(figsize=(20, 14))  # large figure

for i, region in enumerate(regions, start=1):
    plt.subplot(2, 2, i)  # 2 rows x 2 cols layout
    avg_sales_year_df[region].plot(kind='line',
                                   color='#40373E', 
                                   marker = "o",
                                   linewidth=3)
    plt.title(f'Average Sales per Year in {region}')
    plt.xlabel('Year')
    plt.ylabel('Average Sales (millions)')
    plt.xticks(rotation=90)

plt.tight_layout()
plt.show()
```

1. North America (NA)
Peak Years: Late 1980s, especially 1989 with an average above 2.7 million units.
Early Decline: A sharp drop occurred in the early 1990s.
Stagnation: From 1995 to 2015, average sales hovered at a lower but steady level (~0.2–0.4 million).
Recent Dip: Sales continued to decline into the 2010s, with 2020 showing a slight uptick.
North America was a dominant market in the 1980s and early 90s. However, average sales plateaued after that, possibly due 
to market saturation and competition.

2. Europe (EU)
Slow Rise: Growth began in mid-1980s and peaked in 1990–1991, with ~0.5 million average sales.
More Stable than NA: EU maintained consistent average sales from mid-90s onward.
Modern Recovery: 2012–2014 showed a moderate rise, with 2014 nearing peak levels again.
Europe became increasingly important over time, showing less volatility than North America and experiencing sustained growth into the 2010s.

3. Japan (JP)
Early Peak: Japan saw high averages from 1983–1990, peaking above 1.0 million in 1989–1990.
Long Decline: Sales have steadily declined ever since, especially post-2000.
Recent Years: Average sales in 2010s dropped to below 0.1 million.
Insight: Japan was a leader in the early console era (Nintendo, Sega). Over time, local market contraction and global shifts reduced its dominance.

4. Other Regions
Late Start: Significant growth didn't begin until the late 1990s.
Moderate Peaks: Highest averages were in 2008–2014, peaking near 0.09 million.
More Resilient in 2010s: Sales remained relatively high compared to Japan during this period.
"Other" regions (e.g., Latin America, Middle East, Australia) saw delayed but steady adoption, especially during the global
expansion of gaming in the 2000s.

1: Game Genre Trends - 
Certain genres (like Action or Shooter) dominate sales in specific years, reflecting shifts in player preferences and blockbuster releases.

2: Publisher Dominance - 
A few publishers, such as Nintendo, EA, and Activision, consistently lead global sales due to strong IPs and massive franchise successes.

3: Best-Selling Individual Games - Games like Wii Sports, GTA V, and Mario Kart significantly outperform others, indicating that franchises and innovative gameplay are key drivers of extreme success.

4: Platform Influence - Platforms with larger install bases (e.g., PlayStation 2, Wii, DS) have the highest total game sales globally, showing that platform popularity directly influences software performance.

5: Global Genre Preferences - Globally, Action and Sports genres have the highest cumulative sales, likely due to their replayability and wide demographic appeal.

6: Sales Over Time - Total global video game sales have fluctuated year to year, often peaking during major console release years or unique global events like the pandemic.

7: Regional Market Power - North America, Europe, and Japan are the top three regions by revenue, highlighting their long-standing roles as major contributors to the global gaming economy.
