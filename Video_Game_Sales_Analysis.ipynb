```python
avg_sales_per_genre_in_na = video_game_sale.groupby('Genre')['North_America_Sales'].mean()
```

```python
avg_sales_per_genre_in_na
```

```python
avg_sales_per_genre_in_europe = video_game_sale.groupby('Genre')['Europe_Sales'].mean()
```

```python
avg_sales_per_genre_in_europe
```

```python
avg_sales_per_genre_in_japan = video_game_sale.groupby('Genre')['Japan_Sales'].mean()
```

```python
avg_sales_per_genre_in_japan
```

```python
avg_sales_per_genre_in_other = video_game_sale.groupby('Genre')['Other_Sales'].mean()
```

```python
avg_sales_per_genre_in_other
```

```python
avg_sales_df = pd.DataFrame({
    'North America': avg_sales_per_genre_in_na,
    'Europe': avg_sales_per_genre_in_europe,
    'Japan': avg_sales_per_genre_in_japan,
    'Other': avg_sales_per_genre_in_other,
})
```

```python
regions = ['North America', 'Europe', 'Japan', 'Other']

plt.figure(figsize=(15, 12))  

for i, region in enumerate(regions, start=1):
    plt.subplot(2, 2, i)  
    avg_sales_df[region].sort_values(ascending=False).plot(kind='bar', color='#6B3333')
    plt.title(f'Average Sales per Genre in {region}')
    plt.xlabel('Genre')
    plt.ylabel('Average Sales (millions)')
    plt.xticks(rotation=90)

plt.tight_layout()
plt.show()
```

North America (NA): Platform and Shooter genres dominate, while Strategy and Adventure underperform.

Europe (EU): Shooter leads, followed by Platform and Racing; Strategy and Adventure trail behind.

Japan (JP): Role-Playing is most popular, with low interest in Shooters and Adventure games.

Other Regions: Shooter and Racing top the charts, while Strategy, Adventure, and Puzzle lag.

```python
#Genre of Game has Sold the Most in a Single Year
year_sale_group = video_game_sale.groupby(by=["Year", "Genre"])["Global_Sales"].sum().reset_index()
yearly_sales = year_sale_group.groupby(by=["Year"])["Global_Sales"].transform(max) == year_sale_group["Global_Sales"]
yearly_sales_max = year_sale_group[yearly_sales].reset_index(drop = True)
```

```python
genre = yearly_sales_max["Genre"]
```

```python
yearly_sales_max
```

```python
plt.figure(figsize=(30, 15))
g =sns.barplot(x = "Year",
                y = "Global_Sales",
                data = yearly_sales_max,
                palette="Set2")
index = 0
for value in yearly_sales_max['Global_Sales']:
    g.text(index, value + 1, str(genre[index] + '----' +str(round(value, 2))), color='#000', size=14, rotation= 90, ha="center")
    index += 1
plt.xticks(size = 15, rotation = 90)
plt.title("Yearly Global Sales with Top Genre", fontsize=20)
plt.xlabel("Year", fontsize=16)
plt.ylabel("Global Sales", fontsize=16)
plt.tight_layout()
plt.show()
```

### Genre of Game Sold the Most in a Single Year
Action games consistently led global sales from 2004 to 2016, peaking above 130 million units.
Sports games were top sellers in the 1990s and early 2000s, showing their historical popularity.
The sales spike aligns with the high release volumes seen in the first chart, especially around 2008â€“2010.
Post-2016, sales plummet dramatically, echoing the reduction in game releases.
Despite genre diversity in earlier years, Action games dominate sales performance during the industry's peak.

```python
# Top publisher by Count each year
top_publisher = video_game_sale[["Year", "Publisher"]]
top_publisher_data = top_publisher.groupby(by = ["Year", "Publisher"]).size().reset_index(name = "count")
top_publisher_data2 = top_publisher_data.groupby(by = ["Year"])["count"].transform(max) == top_publisher_data["count"]
top_publisher_count = top_publisher_data[top_publisher_data2].reset_index(drop = True)
top_publisher_count = top_publisher_count.drop_duplicates(subset=["Year", "count"], keep = "last").reset_index(drop = True)
```
